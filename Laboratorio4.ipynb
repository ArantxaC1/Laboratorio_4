```python
# Problema 4: Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).
```

```python
import numpy as np

# Creo los arrays "a" y "b"
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calculo la diferencia
dif = a - b

# Calculo la media aritmética
media_a = np.mean(dif)

print("La Media Aritmética de la diferencia es:", media_a)
```

```python
# Problema 5: Tengo valores de cordenadas (x, y) en las columnas
#1.33 4.5
#30.0 10.7
#70.2 0.5
# Agregue a este array las coordenadas (37.1, -3.6). Muestre en pantalla las dimensiones del nuevo array.
```

```python
import numpy as np

def agregar_coordenada(arr_original, nuevo_coord):
  
    # Verificamos que array original tenga 2 columnas
    if arr_original.ndim != 2 or arr_original.shape[1] != 2:
        raise ValueError("debe tener exactamente 2 columnas (x, y).")
    
    # Verificamos que la nueva coordenada tenga 2 elementos
    if not isinstance(nuevo_coord, (list, tuple, np.ndarray)) or len(nuevo_coord) != 2:
        raise ValueError("La coordenada debe tener exactamente 2 elementos: [x, y].")
    
    nuevo_coord = np.array([nuevo_coord])
    nuevo_array = np.vstack([arr_original, nuevo_coord])
    
    print("Nuevo array con coordenadas:")
    print(nuevo_array)
    print("Dimensiones del nuevo array:", nuevo_array.shape)

# Realizamos el array y le agregamos las coordenadas
coord = np.array([
    [1.33, 4.5],
    [30.0, 10.7],
    [70.2, 0.5]
])

nueva = [37.1, -3.6]

agregar_coordenada(coord, nueva)
```

```python
# Problema 2: Realice una gráfica de barras sobre la población de cada ciudad, ordenela de menor a mayo y responda:
# a) ¿Cuál es la segunda ciudad más poblada?
# b) ¿Qué posición ocupa Granada en el ranking delas ciudades más pobladas?
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Agrupamos por municipio y su población
poblacion_por_municipio = df.groupby('Municipio')['Poblacion'].sum()
poblacion_ordenada = poblacion_por_municipio.sort_values() #ordenamos de menor a mayor

# Gráfico de barras
plt.figure(figsize=(12, 6))
poblacion_ordenada.plot(kind='bar', color='purple')

plt.title("Población por Municipio de Menor a Mayor")
plt.ylabel("Habitantes")
plt.xlabel("Municipio")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

#a) segunda ciudad mas poblada
segunda_ciudad = poblacion_ordenada.sort_values(ascending=False).index[1]
print("Segunda ciudad más poblada:", segunda_ciudad)

#b) posicion de Granada en el ranking de ciudades mas pobladas
ranking = poblacion_ordenada.sort_values(ascending=False).reset_index()
granada = ranking[ranking['Municipio'] == 'Granada'].index[0] + 1

print("Granada está en la posición", granada, "del ranking de ciudades más pobladas.")
```

```python
# Problema 6: Realice un histograma con la población de los Municipios para cada una de las provincias
```

```python
import matplotlib.pyplot as plt
poblacion = df.groupby(['Provincia', 'Municipio'])['Poblacion'].sum().reset_index()
provincias = poblacion['Provincia'].unique()

# Histograma por provincia
for provincia in provincias:
    datos_por_provincia = poblacion[poblacion['Provincia'] == provincia]['Poblacion']
    
    plt.figure(figsize=(8, 5))
    plt.hist(datos_por_provincia, bins=10, color='Pink', edgecolor='Purple')
    plt.title(f"Histograma de población - {provincia}")
    plt.xlabel("Población de municipios")
    plt.ylabel("Frecuencia")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# Problema 7: Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?
```

```python
import pandas as pd
import numpy as np

municipios = df.groupby('Municipio')['Poblacion'].sum().reset_index()

# 50 municipios aletorios
muestra_aleatoria = municipios.sample(n=50, random_state=42)

# Calculamos promedio y desviacion
prom = muestra_aleatoria['Poblacion'].mean()
desv = muestra_aleatoria['Poblacion'].std()

# Municipios seleccionados
print("Municipios seleccionados:")
print(muestra_aleatoria.sort_values(by='Poblacion', ascending=False).to_string(index=False))

print("Promedio de población:", prom)
print("Desviación estándar:", desv)
```

```python
# Problema 8: Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero
```

```python
import numpy as np
coord_madrid = df[df['Municipio'] == 'Madrid'][['X', 'Y']].iloc[0]

# Se excluye a Madrid
df_sin_madrid = df[df['Municipio'] != 'Madrid'].copy()

# calculo de distancia
df_sin_madrid['Distancia_a_Madrid'] = np.sqrt((df_sin_madrid['X'] - coord_madrid['X'])**2 + 
                                               (df_sin_madrid['Y'] - coord_madrid['Y'])**2)

# Municipio más cercano y más lejano
mas_cercano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmin()]
mas_lejano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmax()]

print("Municipio más cercano a Madrid:")
print("Nombre:", mas_cercano['Municipio'])

print("Municipio más lejano a Madrid:")
print("Nombre:", mas_lejano['Municipio'])
```
